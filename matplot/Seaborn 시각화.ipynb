# 02. 확률 밀도 함수 (PDF) 
## Probability Density Function 
### 확률 밀도 함수란 히스토그램의 막대수(bins)를 무한할떄랑 같은 그래프
### 확률밀도함수 그래프의 넓이는 '1'이다! 
### 확률 밀도 함수는 데이터셋의 분포를 나타냄 
### 특정 구간의 확률은 그래프 아래 구간의 면적이다.
### 선의 넓이 = 0  ex) 173.5일 확률은 0

### 특정 값에 대한 확률을 구할 수 있는 경우 : 주사위 숫자/동전 개수 
### 특정 범위에 대한 확률을 구할 수 있는 경우 : 키/몸무게

## seaborn import

```python
import pandas as pd 
import seaborn as sns #보통 seaborn 불러올떄 sns라고 함 
```

```python
body_df = pd.read_csv('./data/body.csv', index_col = 0)
body_df.head()
```

```python
body_df['Height']
```

```python
#value_counts()로 개수세주고, sort_index로 정렬해줌
body_df['Height'].value_counts().sort_index()
```

```python
body_df['Height'].value_counts().sort_index().plot()
#아래의 그림을 보면 선그래프는 복잡해보인다.
#-> seaborn으로 kde plot으로 만들어보자.
```

```python
sns.kdeplot(body_df['Height'])
```

```python
#조금더 정확하게 그려주려면 bw= N으로 만들어준다.
#bw = bin width 
#bw : 크기를 줄이면 그래프가 정교해짐, 크기를 키우면 그래프가 부드럽지만 정확도는 떨어짐 
sns.kdeplot(body_df['Height'], bw = 0.5)
```

# 09. 카테고리별 시각화


```python
import pandas as pd
import seaborn as sns 

laptops_df = pd.read_csv('./data/laptops.csv')
laptops_df.head()
```

```python
laptops_df['os'].unique() #os에 대해 종류를 카테고리화 시키는 방법

#sns.catplot()을 사용
sns.catplot(data= laptops_df, x= 'os', y = 'price', kind ='box')
#맥이 가장 가격이 높고, 리눅스가 가장 싸다. 
```

```python
sns.violinplot(data= laptops_df, x= 'os', y = 'price', kind ='box')
```

```python
#strip plot = default : kind 안쓰면 strip plot으로 보여줌 
sns.catplot(data= laptops_df, x= 'os', y = 'price', kind ='strip')
```

```python
laptops_df['processor_brand'].unique()
```

```python
#색깔로 원하는 것을 구분해주는 기능 hue = 'columns명'
sns.catplot(data= laptops_df, x= 'os', y='price', kind = 'strip', hue = 'processor_brand' )

#amd 프로세서가 대체적으로 싼 노트북에 들어간다.
```

```python
sns.catplot(data= laptops_df, x= 'os', y='price', kind = 'swarm', hue = 'processor_brand' )
#모이는게 보기가 어려우면 kind = swarm을 쓰면된다. 
```
