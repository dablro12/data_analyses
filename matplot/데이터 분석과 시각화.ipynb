# 08. 실리콘 밸리에는 누가 일할까? II

어도비 전체 직원들의 직군 분포를 파이 그래프로 그려보세요.

(인원이 0인 직군은 그래프에 표시되지 않아야 합니다.)



```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('./data/silicon_valley_details.csv')

ifsen = ((df['company']=='Adobe') & (df['count'] != 0) & (df['race']=='Overall_totals') & (df['job_category']!='Totals') & (df['job_category']!='Previous_totals')) 
adobe_df = df.loc[ifsen].set_index('job_category')
adobe_df.tail(30)
```
