# 1. 머신 러닝이란?

## 1-1 머신러닝이란?

Machine learning = 기계 학습. 

일반적인 프로그램은 우리가 딱 정해준 대로 업무를 진행함
머신 러닝 프록르맴은 점점 발전해서 그 업무를 더 잘해짐 . 스스로 업무능력을 키움 

예를 들어 스팸 메일을 분류하는 프로그램을 만든다고 가정했을떄, 
일반적인 프로그램은 if문과 문자열 처리로 특정 광고성 문구나 특정 기업의 이름이 연속으로 나올떄 스팸 메일로 처리
-> 이 방식은 인간이 패턴을 분석해서 컴퓨터에 조건문으로 규칙을 정해준 것. 
-> 그러나 이러한 방식은 복잡한 규칙을 캐치하기에 어려움이 있다. -> 완벽하게 분류할 수 없다. -> 크게 개선 어려움
> 이를 해결해주는 것이 머신 러닝! 

머신러닝은 컴퓨터가 직접 이 규칙들을 찾아 낼 수 있도록 하는 것! -> 어떻게? : 데이터(=경험)를 이용해서!
게속해서 데이터를 입력해주어 확률적으로 80% -> 90% -> 100%에 가까운 확률로 규칙을 찾음

> ###  머신러닝 구체적인 정의 : 기계가 학습한다는 건, 프로그램이 특정 작업(T)를 하는데 있어 경험(E)을 통해 작업의 성능(P)을 향상시키는 것 

## 1-2 머신 러닝이 핫해진 이유

머신러닝은 수십 년전부터 연구해온 학문이나 발전은 최근들어서 많아짐 why? data가 많아졌기 때문이다~!
최근들어 데이터가 많이 쌓이고 저장공간이 늘어나면서 데이터로 기록될 수 있게 됨 -> 성능이 좋은 컴퓨터가 계산하기 더 빠름

> ### 1. 머신러닝은 데이터가 충분하지 않으면 성능을 향상시키기가 어렵다! -> 데이터가 충분해야한다!

> ### 2. 머신러닝은 컴퓨터 성능이 좋아야한다!

> ### 3. 활용성이 증명되었다. -> 유튜브와 같은 알고리즘에도 적용하게 되었다!



## 1-3  인공 지능? 빅 데이터? 머신 러닝?

머신러닝 - 빅데이터 - 딥러닝 -인공 지능에 대한 정확한 구분

1. 빅데이터 : 엄청나게 많은 양의 데이터를 다루는 분야
데이터 보관/처리법, 데이터 분석 방법들 

2. 인공지능 : 프로그램이 인간처럼 생각하게 하는 학문

3. 인공지능의 달성을 위해 머신 러닝을 사용하는 것

4. 딥러닝 : 머신 러닝 기법 중 하나로써, 프로그램을 학습시킬 때 복잡한 인공신경망(다중 Layer)로 구성된 알고리즘을 사용
